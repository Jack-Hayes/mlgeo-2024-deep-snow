# Create histogram over time to visualize temporal distribution of datasets

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pandas as pd
```

```python
train_subset_dir = '../../data/subsets_v4/train'
test_subset_dir = '../../data/subsets_v4/test'
val_subset_dir = '../../data/subsets_v4/val'
```

```python
def list_dates(subset_dir):
    subset_list = os.listdir(subset_dir)
    date_list = []
    for subset in subset_list:
        date_list.append(pd.to_datetime(subset.split('_')[4]))
    return date_list
```

```python
train_date_list = list_dates(train_subset_dir)
test_date_list = list_dates(test_subset_dir)
val_date_list = list_dates(val_subset_dir)
```

```python
sns.set_theme()
f, ax = plt.subplots(figsize=(5, 2))
sns.histplot(ax=ax, x=train_date_list, binwidth=25, element='step', fill='cornflowerblue', color='cornflowerblue', alpha=0.7, label='training')
sns.histplot(ax=ax, x=val_date_list, binwidth=25, element='step', fill='orange', color='orange', alpha=0.4, label='validation')
sns.histplot(ax=ax, x=test_date_list, binwidth=25, element='step', fill='purple', color='darkviolet', alpha=0.3, label='testing')
ax.set_ylabel('subset count')
ax.set_xlabel('time')
plt.legend(loc='upper left')
plt.tight_layout()
plt.savefig('../../figs/subset_v4_temporal_distribution.png', dpi=300)
```
